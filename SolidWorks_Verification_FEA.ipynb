The SolidWorks Analysis shows 3 point bending on a rectangular prism. The displacement due to force is shown in red, approximately 7.7 * 10^-4 mm, and the formula to calculate this displacement due to bending is:

<br/>
$$
\delta_{\text{max}} = \frac{P L^3}{48 E I}
$$

```python
import numpy as np

solidworks_displacement = 7.7 * (10 ** (-4)) # mm
P = 18 # N
L = 95 # mm
E = 2011.5484132732429 # N/mm^2
I = (1/12) * (50 ** 3) * (20) # mm^4

max_displacement = (P * (L ** 3)) / (48 * E * I)
print(f"Displacement due to bending is: {max_displacement} mm")
```

```python
percent_diff_displacement = (solidworks_displacement - max_displacement)/((solidworks_displacement + max_displacement)/2) * 100
print(f"Percentage difference in displacement: {percent_diff_displacement} %")
```

```python
solidworks_stress = 2.7 * (10 ** (6))
T = 50 # N m
r = 0.023 # m
J = (np.pi * (r ** 4))/2 # m^4

stress = (T * r) / J
print(f"Stress due to torsion is: {stress} N/m^2")
```

The SolidWorks Analysis shows torsion from a moment on a cylinder. The stress due to torsion is shown in yellow, approximately 2.7 * 10^(6) N/m^2, and the formula to calculate this stress due to torsion is:

<br/>
$$
\tau = \frac{T r}{J}
$$

```python
percent_diff_torsion_stress = (solidworks_stress - stress)/((solidworks_stress + stress)/2) * 100
print(f"Percentage difference in torsion stresses: {percent_diff_torsion_stress} %")
```
